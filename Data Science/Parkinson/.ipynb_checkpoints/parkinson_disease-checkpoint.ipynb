# <center><bold> Modelo de clasificación para la enfermedad de Parkinson </bold></center>

### Entre las afecciones que se generan de la enfermedad de Parkinson (EP), destaca la disartria, la cuál impide gesticular correctamente las palabras al hablar, pues el paciente se ve imposibilitado de controlar correctamente los músculos involucrados en dicha función. La disartría puede estar siendo el resultado de las afecciones de una o más regiones subcorticales implicadas en funciones motoras, como lo són el cerebelo, el tronco encefálico o los ganglios basales.

### Las personas que padecen esta enfermedad suelen hablar de forma lenta, con un tono más monótono, o con menos fuerza en la voz, pero por sobre todo con muchas impresiciones al momento de articular, lo que genera temblores y tartamudeo.

### El conjunto de datos que se presenta a continuación se compone de 195 registros, algunos de ellos corresponden a personas que presentan EP y otros que no. La definición y clasificación de las variables o columnas se presenta a continuación:

- name - Nombre del sujeto ASCII y número de grabación
-   MDVP:Fo(Hz) - Frecuencia fundamental vocal promedio
-   MDVP:Fhi(Hz) - Frecuencia fundamental vocal máxima
-   MDVP:Flo(Hz) - Frecuencia fundamental vocal mínima
-   MDVP:Jitter(%),MDVP:Jitter(Abs),MDVP:RAP,MDVP:PPQ,Jitter:DDP - Varias medidas de variación en la frecuencia fundamental
-    MDVP:Shimmer,MDVP:Shimmer(dB),Shimmer:APQ3,Shimmer:APQ5,MDVP:APQ,Shimmer:DDA - Varias medidas de variación en amplitud.
-    NHR,HNR - Dos medidas de relación de ruido a componentes tonales en la voz.
-    status - Estado de salud del sujeto (uno) - Parkinson, (cero) - saludable
-    RPDE,D2 - Dos medidas de complejidad dinámica no lineal
-    DFA - Exponente de escala fractal de señal
-    spread1,spread2,PPE - Tres medidas no lineales de variación de frecuencia fundamental

```python
classified_variables = { "ID":  ["name"],
     "Medidas de frecuencia vocal":  ["MDVP:Fo(Hz)", "MDVP:Fhi(Hz)", "MDVP:Flo(Hz)"],
     "Medidas de variación en la frecuencia vocal": ["MDVP:Jitter(%)", "MDVP:Jitter(Abs)", "MDVP:RAP", "MDVP:PPQ", "Jitter:DDP"],
     "Medidas de variación en la amplitud vocal": ["MDVP:Shimmer","MDVP:Shimmer(dB)", "Shimmer:APQ3","Shimmer:APQ5","MDVP:APQ","Shimmer:DDA"],
     "Medidas de ruido en la voz" : ["NHR","HNR"],
     "Medidas de complejidad dinámica no lineal" : ["RPDE","D2"],
     "Exponente de escala fractal de señal" : ["DFA"],
     "Medidas de variación en la frecuencia vocal no lineales": ["spread1","spread2","PPE"]
}

x = [1,2,4,6,0,2,4,6]
y = [9,2,2,2,6,6,6,6]
s = []
n = []

for classification, values in classified_variables.items():
    s.append(8000*len(values))
    n.append(classification)
    
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 12.5)

for i, txt in enumerate(n):
    fontsize = 20 -len(txt)/5
    ax.annotate(txt, (x[i], y[i]), fontsize=fontsize)
    
print(len(s))
plt.ylim([0, 10])
plt.xlim([0, 10])
plt.scatter(x,y,s=s, facecolor='blue', alpha=0.2)
plt.show()
```

```python
TARGET = 'status'
FEATURES = [col for col in df.columns if col != TARGET]
RANDOM_STATE = 12

#X = df.drop(TARGET, axis = 1)

X = df[['avg_freq','spread1', 'spread2', 'D2', 'PPE']]
y = df[TARGET]

X
```

```python

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    random_state=12,
                                                    test_size= 0.33)
clf = LazyClassifier(verbose=0,
                     ignore_warnings=True,
                     custom_metric=None,
                     predictions=False,
                     random_state=12,
                     classifiers='all')

models, predictions = clf.fit(X_train, X_test, y_train, y_test)
#clear_output()
```
